Dados: 40 52 55 60 70 75 86 90 90 90 94 94 95 98 100 115 125 150

Determine as 5 estatísticas: min, max, Q1, mediana, Q3.

Interprete os resultados.

min=40; Q1=70; Md=90; Q3=98; max=150

A distribuição dos dados é simétrica? não

Existe algum valor possível/provável que pode ser avaliado como outlier?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dados
dados = np.array([40, 52, 55, 60, 70, 75, 86, 90, 90, 90, 94, 94, 95, 98, 100, 115, 125, 150])

# estatísticas
min_val = np.min(dados)
q1 = np.percentile(dados, 25)
mediana = np.median(dados)
q3 = np.percentile(dados, 75)
max_val = np.max(dados)

min_val, q1, mediana, q3, max_val
```

```python
print("Min:", min_val)
print("Q1:", q1)
print("Mediana:", mediana)
print("Q3:", q3)
print("Max:", max_val)
```

```python
iqr = q3 - q1
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

print("Limite inferior:", limite_inferior)
print("Limite superior:", limite_superior)

# Outliers
outliers = dados[(dados < limite_inferior) | (dados > limite_superior)]
outliers
```

```python
plt.figure(figsize=(14,4))

plt.subplot(1,3,1)
sns.histplot(dados, bins=10, kde=True)
plt.title("Histograma")

plt.subplot(1,3,2)
sns.boxplot(x=dados)
plt.title("Boxplot")

plt.subplot(1,3,3)
sns.violinplot(x=dados)
plt.title("Violin Plot")

plt.tight_layout()
plt.show()
```
